```python
data = pd.read_csv('file.csv') # Заново добавляем файлик
```

```python
data['Avg_Price'].plot(kind='hist', bins=20, title='Распределение цен')
plt.show()
```

```python
data = data.dropna()
```

```python
numeric_data = data.select_dtypes(include=['float64', 'int64'])
correlation = numeric_data.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.show()
```

```python
plt.figure(figsize=(12, 8))  # Ничерта не видно, так что увеличим
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.show()
```

```python
grouped = data.groupby('Product_Category')['Online_Spend'].sum()
print(grouped)

grouped.plot(kind='bar', title='Затраты по категориям товаров')
plt.show()
```

```python
# Сортируем по убыванию
spend_by_category = data.groupby('Product_Category')['Online_Spend'].sum().sort_values(ascending=False)
print(spend_by_category)
```

```python
# Округляем значения, например, до 3 знаков после запятой
spend_by_category_rounded = spend_by_category.round(3)
print(spend_by_category_rounded)
```

```python
# Переводим расходы в миллионы
spend_by_category_in_millions = spend_by_category / 1_000_000
spend_by_category_in_millions_rounded = spend_by_category_in_millions.round(3)
print(spend_by_category_in_millions_rounded)
```

```python
data['Transaction_Date'] = pd.to_datetime(data['Transaction_Date'])
time_series = data.groupby(data['Transaction_Date'].dt.month)['Online_Spend'].sum()
time_series.plot(kind='line', title='Изменение затрат онлайн по месяцам')
plt.show()
```

```python
plt.scatter(data['Avg_Price'], data['Quantity'])
plt.title('Анализ выбросов')
plt.xlabel('Средняя цена')
plt.ylabel('Количество')
plt.show()
```

```python
monthly_spend = data.groupby('Month')['Online_Spend'].sum()
monthly_spend.plot(kind='bar', color='skyblue', title='Online Spend by Month')
plt.xlabel('Month')
plt.ylabel('Total Online Spend')
plt.show()
```

```python
# Пример использования boxplot для нахождения выбросов по 'Online_Spend'
sns.boxplot(data['Online_Spend'])
plt.title('Boxplot for Online Spend')
plt.show()
```

```python
# Вычисляем Q1, Q3 и IQR для столбца Online_Spend
Q1 = data['Online_Spend'].quantile(0.25)
Q3 = data['Online_Spend'].quantile(0.75)
IQR = Q3 - Q1

# Определяем границы для выбросов
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f'Lower Bound: {lower_bound}')
print(f'Upper Bound: {upper_bound}')
```

```python
# Удаляем выбросы из данных
filtered_data = data[(data['Online_Spend'] >= lower_bound) & (data['Online_Spend'] <= upper_bound)]

# Проверим, сколько данных осталось
print(filtered_data.shape)
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=filtered_data['Online_Spend'])
plt.title('Boxplot после удаления выбросов')
plt.show()
```
